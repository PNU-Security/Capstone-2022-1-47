```python
import os
import numpy as np

import nltk
from nltk.tokenize import word_tokenize

import pandas as pd
import subprocess

import collections
from collections import Counter

import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer

import import_ipynb
import makeData as md

#nltk.download('punkt')

DATA_IN_PATH = './data/'


#소문자로 바꾸기, 한글자 제거하기    
def lower_case(scripts):
    for idx, row in scripts.iterrows():
        lower_code = row['code'].lower().split()
        lower_line = []
        for word in lower_code:
            if len(word) >= 2:
                lower_line.append(word)
        scripts.loc[idx, 'code'] = ' '.join(lower_line) 
    return scripts

               
def tokenizing(scripts):
    #Tokenizer 사용 - 빈도수 높을수록 인덱스 번호가 작다
    tokenizer = Tokenizer()  
    tokenizer.fit_on_texts(scripts)
    total_words = tokenizer.word_index
    print("빈도수 높은 단어 10개")
    print(total_words)
    #빈도수 낮은거 제거해보자
    
    #우선 1인거 뽑아볼까 몇개있는지 비율이 어느정도 되는지 어떤게 있는지
    #빈도수 1 제거
    word_frequency = []
    total_count = 0

    for word, count in tokenizer.word_counts.items():
        total_count += count
        if count < 2:
            word_frequency.append(word)
    print("전체 단어 개수 :",len(total_words)) 
    print("빈도수 1인 단어 개수 :",len(word_frequency))
    print("전체 단어에서 빈도수 1인 단어의 비율 : {:.2f}%".format(len(word_frequency)/len(total_words)*100))
    print("빈도수 1인 단어의 등장 빈도 비율 : {:.2f}%".format((len(word_frequency)*2)/total_count*100))

                    
def counter(scripts):
    vocab = Counter()
    
    for script in scripts:
        line = script.split()
        for word in line:
            vocab[word] = vocab[word]+1
        
    vocab_sorted = sorted(vocab.items(), key=lambda x:x[1], reverse=True)
    word_to_index = {}

    i = 1
    total_count = 0
    one_count = 0
    for (word, frequency) in vocab_sorted:
        total_count += frequency
        if frequency > 1:
            word_to_index[word] = i
            i += 1
        else:
            one_count += 1
    print("전체 단어 개수 :",len(word_to_index)) 
    print("빈도수 1인 단어 개수 :",one_count)
    print("전체 단어에서 빈도수 1인 단어의 비율 : {:.2f}%".format(one_count/len(word_to_index)*100))
    print("빈도수 1인 단어의 등장 빈도 비율 : {:.2f}%".format((one_count)/total_count*100))
```
